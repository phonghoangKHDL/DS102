```python
import os
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer, make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.pipeline import make_pipeline
from imblearn.pipeline import make_pipeline as imbl_pipe
from imblearn.over_sampling import SMOTE
```

## Load Data

```python
abt = pd.read_csv("../data/churn_prepared_data.csv")

# Object for target variable
y = abt.Exited

# Object for input features
X = abt.drop(['Exited'], axis=1)

print(X.shape, y.shape)
```

## Separate dataframe into separate object

```python
# List numerical features
num_columns = X.select_dtypes(include='number').columns.tolist()
num_columns
```

```python
# List categorical features
cat_columns = X.select_dtypes(include='object').columns.tolist()
cat_columns
```

## Create a Train Test split

```python
random_state = 10
# Split X and y into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.25,
                                                    random_state=random_state,
                                                    stratify=abt.Exited)

# Print number of observations in X_train, X_test, y_train, and y_test
print(len(X_train), len(X_test), len(y_train), len(y_test))
```

## Pre-processing pipeline

```python
num_features = [] 

for i in num_columns:
    location = X.columns.get_loc(i)
    num_features.append(location)
print(num_features)  
```

```python
cat_features = []

for i in cat_columns:
    location = X.columns.get_loc(i)
    cat_features.append(location)
print(cat_features) 
```

```python
# Define column transformer - Encoding & Scaling
preprocess = make_column_transformer(
    (MinMaxScaler(), num_features),
    (OneHotEncoder(sparse_output=False), cat_features)
)
preprocess
```

## Build Model Pipeline with SMOTE

```python
# Import classifier
import xgboost as xgb

# Define model with pipeline
model_s = imbl_pipe(preprocess,
                  SMOTE(random_state=random_state),
                  xgb.XGBClassifier(use_label_encoder=False, random_state=random_state, eval_metric='logloss'))
model_s
```

```python
# Create the GridSearchCV model
# Create the GridSearch estimator along with a parameter object containing the values to adjust
from sklearn.model_selection import GridSearchCV

param_grid_s = {'xgbclassifier__gamma': [0.5, 0.8, 1],
              'xgbclassifier__max_depth': [3, 4, 5, 6],
              'xgbclassifier__n_estimators': [50, 100, 200],
              'xgbclassifier__learning_rate': [0.01, 0.1, 0.5],
              'smote__sampling_strategy': ['auto', 0.5, 0.8]
        }
grid_s = GridSearchCV(model_s, param_grid_s, verbose=3, cv= 5, scoring='f1', n_jobs=-1)
```

```python
X_train = X_train.values
X_test = X_test.values
grid_s.fit(X_train, y_train)
```

```python
print(grid_s.best_params_)
```

```python
predictions = grid_s.predict(X_test)
print(classification_report(y_test, predictions))
print("Confusion Matrix:")
cm = confusion_matrix(y_test, predictions)
cm = np.around(cm / cm.sum(axis=1)[:, np.newaxis], 2)
print(cm)
```

## Calculate the AUC Score

```python
from sklearn.metrics import roc_auc_score

y_pred_s= grid_s.predict_proba(X_test)[:, 1]

xgb_auc = roc_auc_score(y_test, y_pred_s)

print(f"XGBoost Model (SMOTE) AUC Score: {xgb_auc:.4f}")
```

## Save the model

```python
import joblib

filename = '../models/xgb_smote.sav'
joblib.dump(grid_s, filename)
```

# Build Model Pipeline with CLASS_WEIGHT

```python
weight_ratio = (y_train == 0).sum() / (y_train == 1).sum()
print(weight_ratio)
```

```python
from sklearn.pipeline import Pipeline

# Define model with pipeline
model_w = Pipeline(steps=[
    ('preprocessor', preprocess), 
    ('xgb_weighted', xgb.XGBClassifier(
        eval_metric='logloss',
        scale_pos_weight=weight_ratio,
        subsample=0.8,           
        colsample_bytree=0.8,    
        random_state=random_state 
    ))
])
model_w
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_grid_w = {
    'xgb_weighted__gamma': [0.5, 1],
    'xgb_weighted__max_depth': [3, 4, 5, 6],
    'xgb_weighted__n_estimators': [50, 100],
    'xgb_weighted__learning_rate': [0.1, 0.5, 1],
    'xgb_weighted__min_child_weight': [5, 10, 20],          
}

grid_w = GridSearchCV(model_w, param_grid_w, verbose=3, cv=5, scoring='f1', n_jobs=-1)
```

```python
grid_w.fit(X_train, y_train)
```

```python
print(grid_w.best_params_)
```

```python
predictions = grid_w.predict(X_test)
print(classification_report(y_test, predictions))
print("Confusion Matrix:")
cm = confusion_matrix(y_test, predictions)
cm = np.around(cm / cm.sum(axis=1)[:, np.newaxis], 2)
print(cm)
```

## Calculate the AUC Score

```python
from sklearn.metrics import roc_auc_score

y_pred_w= grid_w.predict_proba(X_test)[:, 1]

xgb_auc = roc_auc_score(y_test, y_pred_w)

print(f"XGBoost Model (Weighted) AUC Score: {xgb_auc:.4f}")
```

## Save the model

```python
filename = '../models/xgb_weighted.sav'
joblib.dump(grid_w, filename)
```
